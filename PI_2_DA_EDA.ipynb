```python
def sumar_columnas(i,f,name):
    # Lista de nombres de columnas a sumar
    columnas_a_sumar = df1.columns[i:f]  # Esto selecciona todas las columnas a partir de la cuarta columna (√≠ndice 5 al 29).

    # Crea una nueva columna con la suma de las columnas seleccionadas
    df1[name] = df1[columnas_a_sumar].sum(axis=1)


sumar_columnas(5,30,"0,256 mbps - 10 mbps")
sumar_columnas(30,40,"11 mbps - 20 mbps")
sumar_columnas(40,50,"21 mbps - 30 mbps")
sumar_columnas(50,58,"31 mbps - 40 mbps")
sumar_columnas(58,63,"41 mbps - 50 mbps")
sumar_columnas(63,67,"51 mbps - 60 mbps")
sumar_columnas(67,73,"61 mbps - 70 mbps")
sumar_columnas(73,76,"71 mbps - 80 mbps")
sumar_columnas(76,80,"81 mbps - 90 mbps")
sumar_columnas(80,83,"91 mbps - 100 mbps")

df1.drop(df1.columns[5:83], axis=1, inplace=True)

df1.head(3)
```

```python
df1_modificado = df1[['Provincia','Otros',
       '0,256 mbps - 10 mbps', '11 mbps - 20 mbps', '21 mbps - 30 mbps',
       '31 mbps - 40 mbps', '41 mbps - 50 mbps', '51 mbps - 60 mbps',
       '61 mbps - 70 mbps', '71 mbps - 80 mbps', '81 mbps - 90 mbps',
       '91 mbps - 100 mbps']].groupby("Provincia").sum()

df1_modificado.reset_index(inplace=True)

df1_modificado
```

```python
# Tomar el primer valor de los registros como etiquetas del eje x
etiquetas_eje_x = df1_modificado['Provincia']

# Tomar los valores del DataFrame para el eje y
valores_eje_y = df1_modificado.iloc[:, 1:].mean(axis=1)  # Calculamos la media de las velocidades de internet para cada provincia

# Crear el gr√°fico de barras
fig, ax = plt.subplots(figsize=(10, 6))

# Dibujar las barras
ax.bar(etiquetas_eje_x, valores_eje_y)

# A√±adir t√≠tulos y etiquetas
ax.set_xlabel("Provincia")
ax.set_ylabel("Accesos por provincia")
ax.set_title("Cantidad de accesos por Provincia")
plt.xticks(rotation=45, ha='right')

# Mostrar el gr√°fico
plt.tight_layout()

# Si utilizas Jupyter Notebook, esto har√° que los gr√°ficos se muestren en l√≠nea.
plt.show()
```

## DataSet: Penetraci√≥n de Internet fijo (accesos por cada 100 hogares)

```python
# Carga de archivos
df3 = pd.read_csv("Dataset/Internet_Penetracion-por-100-hog-provincia.csv") 
df3.head()
```

```python
df3['Accesos por cada 100 hogares'] = df3['Accesos por cada 100 hogares'].str.replace(',', '.')
df3['Accesos por cada 100 hogares'] = df3['Accesos por cada 100 hogares'].astype(float)

df3.head()
```

```python
df3.columns
```

```python
#df3.groupby(['A√±o', 'Trimestre'])['Accesos por cada 100 hogares'].sum()

df3_modificado = df3[['A√±o', 'Trimestre', 'Accesos por cada 100 hogares']].groupby(['A√±o', 'Trimestre']).mean()
df3_modificado.reset_index(inplace=True)
df3_modificado
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que tienes el DataFrame 'df3_modificado' con las columnas 'A√±o', 'Trimestre' y 'Accesos por cada 100 hogares'

# Crear una nueva columna 'Trimestre_num' que combine el a√±o y el trimestre (ejemplo: 2014-Q1)
df3_modificado['Trimestre_num'] = df3_modificado['A√±o'].astype(str) + '-Q' + df3_modificado['Trimestre'].astype(str)

# Establecer el √≠ndice del DataFrame como 'Trimestre_num'
df3_modificado.set_index('Trimestre_num', inplace=True)

# Crear el gr√°fico de l√≠neas
plt.figure(figsize=(10, 6))
plt.plot(df3_modificado.index, df3_modificado['Accesos por cada 100 hogares'], marker='o', linestyle='-', color='b')
plt.xlabel('Trimestre')
plt.ylabel('Accesos por cada 100 hogares')
plt.title('Crecimiento de Accesos por cada 100 hogares por Trimestre')
plt.tight_layout()
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejorar la legibilidad
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que tienes el DataFrame 'df3_modificado' con las columnas 'A√±o', 'Trimestre' y 'Accesos por cada 100 hogares'

# Agrupar por a√±o y sumar los valores de los cuatro trimestres
df3_modificado_anual = df3_modificado.groupby('A√±o')['Accesos por cada 100 hogares'].mean()

# Calcular la variaci√≥n porcentual entre cada a√±o
variacion_porcentual = df3_modificado_anual.pct_change() * 100

# Crear el gr√°fico de l√≠neas
plt.figure(figsize=(10, 6))
plt.plot(df3_modificado_anual.index, df3_modificado_anual.values, marker='o', linestyle='-', color='b', label='Accesos por cada 100 hogares')
plt.xlabel('A√±o')
plt.ylabel('Accesos por cada 100 hogares')
plt.title('Crecimiento de Accesos por cada 100 hogares por A√±o')
plt.tight_layout()
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mejorar la legibilidad
plt.legend()

# A√±adir etiquetas de variaci√≥n porcentual en el gr√°fico
for i in range(1, len(df3_modificado_anual)):
    plt.annotate(f'{variacion_porcentual.iloc[i]:.2f}%', (df3_modificado_anual.index[i], df3_modificado_anual.iloc[i]), 
                 textcoords="offset points", xytext=(-10, 5), ha='center', fontsize=8, color='purple')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supongamos que tienes el DataFrame 'df3' con las columnas 'Provincia' y 'Accesos por cada 100 hogares'

# Calcular la media de accesos por cada 100 hogares para cada provincia
accesos_por_provincia = df3.groupby('Provincia')['Accesos por cada 100 hogares'].mean()

# Ordenar los valores en orden descendente para facilitar la visualizaci√≥n
accesos_por_provincia = accesos_por_provincia.sort_values(ascending=False)

# Crear el gr√°fico de barras
plt.figure(figsize=(12, 6))
plt.bar(accesos_por_provincia.index, accesos_por_provincia.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Provincia')
plt.ylabel('Accesos por cada 100 hogares (Media)')
plt.title('Accesos por cada 100 hogares por Provincia')
plt.tight_layout()
plt.show()
```
